```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
file_path = "twitter_training.csv"
# The dataset doesn't have a header, so we assign names based on common structure
df = pd.read_csv(file_path, header=None, encoding='latin1')
df.columns = ['ID', 'Topic', 'Sentiment', 'Tweet']

print("Initial Data Snapshot:")
print(df.head())
print("\nData Info:")
df.info()
```

```python
# --- 2. Data Preprocessing and Cleaning ---

# Drop the 'ID' and 'Topic' columns as they are not needed for sentiment classification
df = df[['Sentiment', 'Tweet']]

# Drop rows with any missing values (Tweet column is most likely to be missing)
df.dropna(inplace=True)

# Convert Sentiment to numeric labels for easier modeling (Target Variable)
# Mapping: Negative -> 0, Neutral -> 1, Positive -> 2, Irrelevant -> 3
sentiment_mapping = {'Negative': 0, 'Neutral': 1, 'Positive': 2, 'Irrelevant': 3}
df['Sentiment_Label'] = df['Sentiment'].map(sentiment_mapping)

# Check the distribution of the target variable
print("\nSentiment Distribution:")
print(df['Sentiment'].value_counts())
```

```python
# Text Cleaning Function
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Convert to string in case of non-string values
    if not isinstance(text, str):
        return ""
    # Remove mentions (@...) and URLs (http...)
    text = re.sub(r'@[A-Za-z0-9_]+|https?://\S+|www\.\S+', '', text)
    # Remove HTML entities like &amp; &lt; etc.
    text = re.sub(r'&[A-Za-z]+;', ' ', text)
    # Remove special characters, numbers, and punctuation (keeping only letters)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Convert to lowercase
    text = text.lower()
    # Tokenize, remove stopwords, and lemmatize
    tokens = text.split()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words and len(word) > 1]
    return ' '.join(tokens)
```

```python
# Apply preprocessing to the 'Tweet' column
df['Cleaned_Tweet'] = df['Tweet'].apply(preprocess_text)

print("\nSample Cleaned Data:")
print(df[['Tweet', 'Cleaned_Tweet', 'Sentiment']].head())
```

```python
# --- 3. Feature Engineering (Vectorization) ---

X = df['Cleaned_Tweet']
y = df['Sentiment_Label']

# Split data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print(f"\nTraining Samples: {len(X_train)}")
print(f"Testing Samples: {len(X_test)}")

# Initialize TF-IDF Vectorizer
# TF-IDF (Term Frequency-Inverse Document Frequency) assigns a weight to each word based on its frequency in a document and its rarity across all documents.
tfidf_vectorizer = TfidfVectorizer(max_features=10000, ngram_range=(1, 2)) # Use up to 10k features and single/bi-grams

# Fit the vectorizer on the training data and transform both train and test data
X_train_vec = tfidf_vectorizer.fit_transform(X_train)
X_test_vec = tfidf_vectorizer.transform(X_test)

print(f"\nTF-IDF Matrix Shape (Train): {X_train_vec.shape}")
```

```python
# --- 4. Model Training (Logistic Regression) ---

print("\n--- Training Logistic Regression Model ---")
# Logistic Regression is a simple yet powerful linear model for classification
model = LogisticRegression(max_iter=1000, random_state=42, solver='liblinear')
model.fit(X_train_vec, y_train)

print("Training Complete.")
```

```python
# --- 5. Model Evaluation ---

# Predict on the test set
y_pred = model.predict(X_test_vec)

# Calculate Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nModel Accuracy: {accuracy:.4f}")

# Detailed Classification Report
target_names = list(sentiment_mapping.keys())
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=target_names))

# Confusion Matrix
conf_mat = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names, yticklabels=target_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# --- 6. Model Testing with New Data ---

def predict_sentiment(text_list, model, vectorizer, mapping):
    # Preprocess the new text
    cleaned_texts = [preprocess_text(text) for text in text_list]
    # Vectorize the cleaned text
    text_vec = vectorizer.transform(cleaned_texts)
    # Make prediction
    predictions = model.predict(text_vec)
    # Inverse map the label to the original sentiment word
    inverse_mapping = {v: k for k, v in mapping.items()}
    sentiment_results = [inverse_mapping[label] for label in predictions]
    return sentiment_results

new_tweets = [
    "I absolutely love this new game! It's fantastic.",
    "This product is terrible and a complete waste of money.",
    "The news today was neither good nor bad, just facts.",
    "I'm talking about the old phone I used to have, nothing to do with the topic.",
    "I hate this platform"
]

predictions = predict_sentiment(new_tweets, model, tfidf_vectorizer, sentiment_mapping)

print("\n--- Predictions on New Tweets ---")
for tweet, sentiment in zip(new_tweets, predictions):
    print(f"Tweet: **'{tweet}'** -> Predicted Sentiment: **{sentiment}**")
```
